###### Code that Takes a Data Frame and adds a Grade Column

```python
#importing the libraries
import pandas as pd
import numpy as np
# Creating a dict for dataframe
d = {                                                
    'StudentName':['Harrison','Jake','Drake','Hayley','Paul','Dawson','Hamphrey','Gideon','Nyakundi','muia','Caleb','Wilcoxon'],    
    'Score':[64,68,61,86,101,50,-12,39,97,78,120,89.9]}
df = pd.DataFrame(d)   # converting dictionary to dataframe
# apply np.where(), similar to a nested if
df['Grade'] =   np.where((df.Score >= 0)  & (df.Score < 40), 'F',
                np.where((df.Score >= 40) & (df.Score < 50), 'D', 
                np.where((df.Score >= 50) & (df.Score < 60), 'C', 
                np.where((df.Score >= 60) & (df.Score < 70), 'B',
                np.where((df.Score >= 70) & (df.Score < 90), 'A', 
                np.where((df.Score >= 90) & (df.Score <= 100),'A*',
                'Invalid input'))))))
df
```

```python
df=pd.read_csv("C:/Users/USER/Desktop/Statistics $ Programming/Grading system.csv")
dfNew=df.copy()
dfNew
```

```python
df3=dfNew.drop(['GRADE'],axis='columns')
df3
```

```python
df3=dfNew.drop(['GRADE'],axis='columns')
df3['GRADES'] = np.where((df3.AVERAGE >= 0)  & (df3.AVERAGE < 40), 'E',
                np.where((df3.AVERAGE >= 40) & (df3.AVERAGE < 50), 'D', 
                np.where((df3.AVERAGE >= 50) & (df3.AVERAGE < 60), 'C', 
                np.where((df3.AVERAGE >= 60) & (df3.AVERAGE < 70), 'B',
                np.where((df3.AVERAGE >= 70) & (df3.AVERAGE <= 100), 'A', 
                'Invalid input')))))

df3['DIVISION'] = np.where((df3.AVERAGE >= 0) & (df3.AVERAGE < 40), 'FAIL!',
                np.where((df3.AVERAGE >= 40) & (df3.AVERAGE < 50), 'PASS', 
                np.where((df3.AVERAGE >= 50) & (df3.AVERAGE < 60), 'SECOND CLASS LOWER', 
                np.where((df3.AVERAGE >= 60) & (df3.AVERAGE < 70), 'SECOND CLASS UPPER',
                np.where((df3.AVERAGE >= 70) & (df3.AVERAGE <= 100), 'FIRST CLASS!',
                'Invalid input')))))
df3
```

###### Python program to calculate grade of student based on marks obtained by 5 subjects

```python
print('What is your name? ')
Name=(input())
print("")
print('Hi '+ Name +" Please enter marks obtained in each subjects: ")
print("")
print('English: ')
markOne=int(input())
print('Kiswahili: ')
markTwo=int(input())
print('Geography: ')
markThree=int(input())
print('Mathematics: ')
markFour=int(input())
print('Physics:')
markFive=int(input())
Total=markOne+markTwo+markThree+markFour+markFive
Average=Total/5
print("")
print('AVERAGE SCORE: ' + str(Average))
print("")
if Average>=70 and Average <=100:
    print(Name + ' Your mean grade is A') 
    print('FIRST CLASS!')
    print("")
    print("Well Done <(^_^)>")
elif Average>=60 and Average <=69:
    print(Name + ' Your mean grade is B')
    print('SECOND CLASS UPPER DIVISION.')
elif Average>=50 and Average <=59:
    print(Name + ' Your mean grade is C')
    print('SECOND CLASS LOWER DIVISION.')
elif Average>=40 and Average <=49:
    print(Name + ' Your mean grade is D')
    print('PASS.')
elif Average>= 0 and Average <=39 :
    print(Name + ' You Failed!')
    print("")
    print("Pull up your socks! (-_-)")
else:
    print(Name + ' Your marks are irregula!')
```

###### Receive Marks using for Loop and find Grade

```python
mark=[]
total=0
print('Enter Marks Obtained in 5 Subjects: ')
for i in range(5):
    mark.insert(i,input())

for i in range(5):
    total=total+int(mark[i])
avg=total/5
print('AVERAGE SCORE: ' + str(avg))
if avg>=70 and avg <=100:
    print('Your grade is A') 
elif avg>=60 and avg < 70:
    print('Your grade is B')
elif avg>=50 and avg < 60:
    print('Your grade is C')
elif avg>=40 and avg < 50:
    print('Your grade is D')
elif avg>= 0 and avg < 40 :
    print('Your grade is Fail')
else:
    print('Your marks are irregula!')
    

    
```
